```python
# 1. Instalasi Library yang Diperlukan
!pip install opencv-python-headless
!pip install numpy
!pip install scikit-learn

# 2. Import Library
import cv2
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

# 3. Fungsi untuk Ekstraksi Fitur
def extract_features(image):
    # Mengubah citra ke ruang warna HSV
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    h_mean = np.mean(hsv[:, :, 0])
    s_mean = np.mean(hsv[:, :, 1])
    v_mean = np.mean(hsv[:, :, 2])

    # Ekstraksi fitur bentuk (contoh: eccentricity)
    # Anda bisa menambahkan lebih banyak fitur sesuai kebutuhan

    return [h_mean, s_mean, v_mean]

# 4. Mempersiapkan Data
# Upload dataset citra bunga ke Google Colab
from google.colab import files
uploaded = files.upload()  # Upload dataset

# Misalkan kita memiliki folder 'flowers' yang berisi subfolder untuk setiap spesies
# Contoh struktur folder:
# flowers/
# ├── daisy/
# │   ├── image1.jpg
# │   ├── image2.jpg
# ├── rose/
# │   ├── image1.jpg
# │   ├── image2.jpg

# 5. Membaca Dataset dan Ekstraksi Fitur
image_paths = []
labels = []

# Ganti 'flowers' dengan nama folder yang sesuai
base_path = 'flowers'

# Get the list of uploaded files
uploaded_files = list(uploaded.keys())

# Iterate over the uploaded files
for species_file in uploaded_files:
    # Assuming file names are like 'species.jpg'
    species = os.path.splitext(species_file)[0] # extract the species name from file name.

    # Read and process the image
    image_content = uploaded[species_file]
    image_data = np.frombuffer(image_content, np.uint8)
    image = cv2.imdecode(image_data, cv2.IMREAD_COLOR) # read image data from uploaded content.
    if image is not None:
        features = extract_features(image)
        image_paths.append(features)
        labels.append(species)  # Label sesuai dengan spesies

# 6. Membagi Dataset menjadi Data Latih dan Data Tes
X_train, X_test, y_train, y_test = train_test_split(image_paths, labels, test_size=0.2, random_state=42)

# 7. Klasifikasi Menggunakan Naïve Bayes
model = GaussianNB()
model.fit(X_train, y_train)
predictions = model.predict(X_test)

# 8. Evaluasi Model
accuracy = accuracy_score(y_test, predictions)
cm = confusion_matrix(y_test, predictions)

# 9. Menampilkan Hasil
print(f'Akurasi: {accuracy * 100:.2f}%')
print('Confusion Matrix:')
print(cm)

# 10. (Opsional) Menampilkan beberapa gambar dan prediksi
def display_images(images, labels, predictions):
    plt.figure(figsize=(10, 10))
    for i in range(len(images)):
        plt.subplot(5, 5, i + 1)
        plt.imshow(images[i])
        plt.title(f'True: {labels[i]}\nPred: {predictions[i]}')
        plt.axis('off')
    plt.show()

# Menampilkan beberapa gambar dari data tes
# Ganti 'X_test' dengan citra asli jika ingin menampilkan gambar
# Anda perlu menyimpan citra asli dalam list untuk menampilkannya
# display_images(X_test_images, y_test, predictions)  # Uncomment jika Anda memiliki citra asli
```
